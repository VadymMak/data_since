```python
df_close = df['close'].copy()
df_close.head()
```

```python
df_close.dropna().plot(figsize=(15, 8), fontsize=13, logy=True)
plt.grid(True)
plt.legend()
plt.show()

plt.close()
```

```python
# Normalize 

df_min = df_close.dropna().iloc[0]
df_min
```

```python
# Step 3: Normalize each column using its first non-NaN value
norm = df_close.copy()
for col in norm.columns:
    first_value = norm[col].dropna().iloc[0]  # First non-NaN value for this column
    norm[col] = norm[col].div(first_value).mul(100)
```

```python
# Step 4: Fill leading NaN values with 100 (normalized starting point) and forward fill
# First, set all NaN to 100 to handle leading NaN
norm = norm.mask(norm.isna(), 100)
# Then, forward fill to propagate values
norm = norm.ffill()
```

```python
# Step 5: Interpolate any remaining gaps with quadratic method
norm = norm.interpolate(method='quadratic')
```

```python
# Step 6: Data quality checks and cleaning
if norm.index.duplicated().any():
    print("Warning: Duplicate dates found. Dropping duplicates.")
    norm = norm[~norm.index.duplicated()]

norm.dropna()

norm
```

```python
norm.dropna().plot(figsize=(15, 8), fontsize=13, logy=True)
plt.legend()
plt.grid(True)
plt.show()

plt.close()
```
