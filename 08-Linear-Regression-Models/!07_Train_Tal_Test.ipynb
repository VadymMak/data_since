```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_one = Ridge(alpha=100)
```

```python
model_one.fit(X_train, y_train)
```

```python
y_eval_predictions = model_one.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval, y_eval_predictions)
```

```python
model_two = Ridge(alpha=1)
```

```python
model_two.fit(X_train, y_train)
```

```python
new_eval_predictions = model_two.predict(X_eval)
```

```python
mean_squared_error(y_eval, new_eval_predictions)
```

```python
y_final_test_predictions = model_two.predict(X_test)
```

```python
mean_squared_error(y_test, y_final_test_predictions)
```
