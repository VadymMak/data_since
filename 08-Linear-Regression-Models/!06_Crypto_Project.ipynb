```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('crypto.csv', header=[0, 1], index_col=0, parse_dates=[0])
df.head(2)
```

```python
df_close_vol = df[[col for col in df.columns if (col[0] == 'close'or col[0] == 'volume')]]
df_close_vol.head(2)
```

```python
df_close_vol.columns = ['_'.join(col).strip() for col in df_close_vol.columns].copy()
df_close_vol.head(2)
```

```python
df = df_close_vol.copy()
df.head(2)
```

```python
df = df.dropna().copy()
df.head(2)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns.tolist()
```

```python
for crypto in ['ada', 'btc', 'doge', 'eth', 'sol', 'xrp']:
    df[f'log_return_{crypto}'] = np.log(df[f'close_{crypto}'] / df[f'close_{crypto}'].shift(1))
```

```python
df = df.dropna().copy()
df.head(2)
```

```python
# Step 3.1: Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
df.shape
```

```python
# Step 3.3: Create lagged features for log returns
lags = [1, 7, 60]
for crypto in ['ada', 'btc', 'doge', 'eth', 'sol', 'xrp']:
    for lag in lags:
        df[f'log_return_{crypto}_lag{lag}'] = df[f'log_return_{crypto}'].shift(lag)
```

```python
# Drop rows with NaN due to lagging (first 60 rows)
df = df.dropna()
print("\nDataFrame shape after adding lagged features and dropping NaNs:", df.shape)
print("\nColumns after adding lagged features:")
print(df.columns.tolist())
```

```python
# Step 3.4: Define the target variable (60-day ahead BTC price)
df['target_btc_60'] = df['close_btc'].shift(-60)
```

```python
# Alternative: Define the target as 60-day ahead log return
df['target_log_return_btc_60'] = df['log_return_btc'].shift(-60)
```

```python
# Drop rows with NaN in the target (last 60 rows)
df = df.dropna()
print("\nDataFrame shape after adding target and dropping NaNs:", df.shape)
print("\nFirst 5 rows with target columns:")
print(df[['close_btc', 'log_return_btc', 'target_btc_60', 'target_log_return_btc_60']].head(2))
```

```python
df.head(2)
```

```python
df_features = df.drop(['target_btc_60', 'target_log_return_btc_60'], axis=1)
df_features.head(2)
```

```python
X = df_features
y = df['target_btc_60']
y_alt = ['target_log_return_btc_60']
```

```python
# Step 8.3: Perform a time-based train-test split
train_size = int(len(X) * 0.7)  # 70% for training
X_train = X.iloc[:train_size]
y_train = y.iloc[:train_size]
X_test = X.iloc[train_size:]
y_test = y.iloc[train_size:]
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
poly_converter = PolynomialFeatures(degree=2, include_bias=False)
poly_converter.fit(X_train_scaled)

X_train_features = poly_converter.transform(X_train_scaled)
X_test_features = poly_converter.transform(X_test_scaled)
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import TimeSeriesSplit
```

```python
tscv = TimeSeriesSplit(n_splits=5)
model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1], n_alphas=5, max_iter=2000000, cv = tscv, 
                    tol=0.005, random_state=101)

model.fit(X_train_features, y_train)
```

```python
# Step 10.4: Compute predictions
y_train_pred = model.predict(X_train_features)
y_test_pred = model.predict(X_test_features)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Step 10.5: Compute RMSE
train_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores = -cross_val_score(model, X_train_features, y_train, cv=4, scoring='neg_mean_squared_error')
cv_RMSE = np.sqrt(np.mean(cv_scores))
```

```python
print(f'Train RMSE: {train_RMSE: .4f}, Test RMSE: {test_RMSE: .4f}, CV RMSE: {cv_RMSE: .4f}, DIFF: {(test_RMSE - train_RMSE): .4f}')
```
