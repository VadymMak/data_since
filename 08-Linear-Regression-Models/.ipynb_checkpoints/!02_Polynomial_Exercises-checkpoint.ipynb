```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error

# Загрузка данных
df = pd.read_csv('Advertising.csv')
X = df.drop('sales', axis=1)
y = df['sales']

# Разделение данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Полиномиальные признаки для degree=2
poly_converter = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly_converter.fit_transform(X_train)
X_test_poly = poly_converter.transform(X_test)

# Масштабирование
scaler = StandardScaler()
scaler.fit(X_train_poly)
X_train_scaled = scaler.transform(X_train_poly)
X_test_scaled = scaler.transform(X_test_poly)

# Обучение модели
model = ElasticNetCV(
    l1_ratio=[0.5, 1],
    n_alphas=10,
    max_iter=10000,
    tol=0.01,
    cv=3,
    random_state=101,
    alphas=np.logspace(-5, -1, 10)
)
model.fit(X_train_scaled, y_train)

# Предсказания и ошибки
train_pred = model.predict(X_train_scaled)
test_pred = model.predict(X_test_scaled)
train_RMSE = np.sqrt(mean_squared_error(y_train, train_pred))
test_RMSE = np.sqrt(mean_squared_error(y_test, test_pred))

# Кросс-валидационная RMSE
cv_scores = -cross_val_score(model, X_train_scaled, y_train, cv=3, scoring='neg_mean_squared_error')
cv_RMSE = np.sqrt(np.mean(cv_scores))

print(f"Degree=2, Train RMSE={train_RMSE:.4f}, Test RMSE={test_RMSE:.4f}, CV RMSE={cv_RMSE:.4f}, "
      f"Best alpha={model.alpha_:.6f}, l1_ratio={model.l1_ratio_}")

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(degrees, train_rmse_error, label='Train RMSE', marker='o')
plt.plot(degrees, cv_rmse_error, label='CV RMSE', marker='o')
plt.title('Train vs CV RMSE for Different Polynomial Degrees', fontsize=14)
plt.xlabel('Degree', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
