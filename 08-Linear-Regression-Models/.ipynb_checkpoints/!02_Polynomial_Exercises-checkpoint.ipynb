```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error
```

```python
# Инициализация списков и разбиение данных
train_rmse_error = []
test_rmse_error = []
cv_rmse_error = []
degreeses = range(1, 11)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)


# Цикл по степеням
for d in degreeses:
    # Полиномиальные признаки
    poly_converter = PolynomialFeatures(degree=d, include_bias=False)
    X_train_poly = poly_converter.fit_transform(X_train)
    X_test_poly = poly_converter.transform(X_test)  # Используем transform для тестовых данных

    # Масштабирование
    scaler = StandardScaler()
    scaler.fit(X_train_poly)
    X_train_scaled = scaler.transform(X_train_poly)
    X_test_scaled = scaler.transform(X_test_poly)

    # Обучение модели
    model = ElasticNetCV(
        l1_ratio=[.1, .5, .7, .9, .95, .99, 1],
        n_alphas=10,
        max_iter=1000000,
        tol=0.01,
        cv=3,
        random_state=101,
        alphas=np.logspace(-5, -1, 10)
    )

    model.fit(X_train_scaled, y_train)

    # Предсказания и ошибки
    train_pred = model.predict(X_train_scaled)
    test_pred = model.predict(X_test_scaled)

    train_RMSE = np.sqrt(mean_squared_error(y_train, train_pred))
    test_RMSE = np.sqrt(mean_squared_error(y_test, test_pred))

    # Кросс-валидационная RMSE
    cv_scores = -cross_val_score(
        model,
        X_train_scaled,
        y_train,
        cv=3,
        scoring='neg_mean_squared_error'
    )
    cv_RMSE = np.sqrt(np.mean(cv_scores))

    # Сохранение ошибок
    train_rmse_error.append(train_RMSE)
    test_rmse_error.append(test_RMSE)
    cv_rmse_error.append(cv_RMSE)

    # Печать результатов
    rmse_diff = test_RMSE - train_RMSE
    print(f'degree: {d}, Train RMSE: {train_RMSE:.4f}, Test RMSE: {test_RMSE:.4f}, CV RMSE: {cv_RMSE:.4f}, '
          f'Best alpha: {model.alpha_:.6f}, l1_ratio: {model.l1_ratio_}, RMSE_DIFF: {rmse_diff:.6f}')

# Построение графика
plt.figure(figsize=(16, 6))
plt.plot(degreeses, train_rmse_error, color='red', label='Train RMSE', marker='o')
plt.plot(degreeses, test_rmse_error, color='blue', label='Test RMSE', marker='o')
plt.plot(degreeses, cv_rmse_error, color='green', label='CV RMSE', marker='o')

plt.title('RMSE for Different Polynomial Degrees')
plt.xlabel('Degrees')
plt.ylabel('RMSE')
plt.grid(True, linestyle='--', alpha=0.9, color='grey')
plt.legend()

plt.show()
```
