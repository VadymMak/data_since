```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Enhanced sample dataset
np.random.seed(42)  # For reproducibility
dates = pd.date_range(start='2025-01-01', periods=200, freq='D')
data = {
    'Date': dates,
    'Sales_Amount': np.random.normal(500, 100, 200),
    'Discount_Percent': np.random.uniform(0, 20, 200),
    'Customer_Segment': np.random.choice(['New', 'Returning', 'Loyal'], 200),
    'Store_ID': np.random.choice(['S001', 'S002', 'S003', 'S004'], 200),
    'Promotion_Active': np.random.choice([True, False], 200, p=[0.6, 0.4]),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 200)
}
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)
```

```python
pivot_df = df.pivot_table(
    values='Sales_Amount', 
    index='Promotion_Active', 
    columns='Region',
    aggfunc='sum'
)

pivot_df
```

```python
plt.figure(figsize=(16, 6), dpi=150)
sns.heatmap(data=pivot_df, annot=True, fmt='.2f', annot_kws={'size': 12})
plt.xticks(rotation=45, fontsize=16)
plt.xlabel('Region', fontsize=20)
plt.yticks(rotation=45, fontsize=16)
plt.ylabel('Promotion Active', fontsize=20)
plt.title('Pivot data frame', fontsize=22)

plt.show()
```

```python
resampled_df = df.resample('ME').agg({
    'Sales_Amount': 'mean'
})
resampled_df
```

```python
plt.figure(figsize=(16, 6), dpi=150)
sns.barplot(
    data=resampled_df, 
    x=resampled_df.index,
    y=resampled_df['Sales_Amount'], 
    hue=resampled_df.index
)
plt.xticks(rotation=45, fontsize=14)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Sales amount', fontsize=18)
plt.title('Monthly Sales Amount visualization', fontsize=22)

plt.show()
```

```python
grouped_df = df.groupby(['Region', 'Customer_Segment']).agg({
    'Sales_Amount': 'sum',
    'Discount_Percent': 'mean'
})

grouped_df
```

```python
unstacked_df = grouped_df['Sales_Amount'].unstack(level='Region')
unstacked_df
```

```python
plt.figure(figsize=(16, 6), dpi=150)
sns.heatmap(
    data=unstacked_df, 
    annot=True, 
    annot_kws={'size': 16}, 
    fmt='.2f'
)
plt.xticks(rotation=45, fontsize=14)
plt.xlabel('Region', fontsize=18)
plt.yticks(rotation=45, fontsize=14)
plt.ylabel('Customer Segment', fontsize=18)
plt.title('Customer segment distribution by region', fontsize=24)

plt.show()
```
